Weighted Trees and Distances

```python
T =graphs.RandomTree(6)

for e in T.edges():
    T.set_edge_label(e[0], e[1], randint(1,10))
    
T.show(edge_labels=True)

```

```python
T.distance(0,2, by_weight=True)
```

```python
T.distance(0,2)
```

```python
D = T.distance_matrix(by_weight=True)
D
```

```python
leaves = [v for v in T.vertices() if T.degree(v) == 1]
D_leaves = D.matrix_from_rows_and_columns(leaves, leaves)
D_leaves
```
