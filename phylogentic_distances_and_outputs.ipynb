Each entry of D_leaves represents the graph distance between two leaves in the tree.
We restrict to leaves because they correspond to the observed entities of interest, rather than internal ancestral nodes.
In phylogenetics, such a matrix can arise from a tree whose leaves represent species, with the distances encoding evolutionary divergence.
